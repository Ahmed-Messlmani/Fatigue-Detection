```
import cv2
import math

videoFile = "/content/drive/Shared drives/Drowsy Driving Dataset/DrowsyDrivingDataset/Watching_Simulation/219.AVI"
imagesFolder = "/content/drive/My Drive/drowsy/video19"
cap = cv2.VideoCapture(videoFile)
frameRate = cap.get(5) #frame rate
while(cap.isOpened()):
    frameId = cap.get(1) #current frame number
    ret, frame = cap.read()
    if (ret != True):
        break
    if (frameId % math.floor(frameRate) == 0):
        filename = imagesFolder + "/video19" +  str(int(frameId)) + ".jpg"
        cv2.imwrite(filename, frame)
    print('Saved image ', frameId)    
cap.release()
```
